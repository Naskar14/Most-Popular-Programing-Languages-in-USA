```python
with open('C:/Users/koushik.n/language.csv') as languages:
    language = csv.DictReader(languages)
    
    language_counter = Counter()
    
    for row in language:
        language_counter.update(row['LanguageWorkedWith'].split(';'))
        
lang = []
popularity = []

for items in language_counter.most_common(15):
    lang.append(items[0])
    popularity.append(items[1])
        
print(lang)
print(popularity)
```

```python
pl.barh(lang,popularity)


pl.title('Most Popular Programing Language in USA')
pl.xlabel('No of People working with')
pl.ylabel('Programing Languages')
pl.show()
```

```python
pl.barh(lang,popularity)

lang.reverse()
popularity.reverse()

pl.title('Most Popular Programing Language in USA')
pl.xlabel('No of People working with')
pl.ylabel('Programing Languages')
pl.grid(True)
pl.show()
```
